```python
import numpy as np
import pandas as pd
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt 
import statsmodels.graphics.tsaplots as sgt 
import statsmodels.tsa.stattools as sts 
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns
sns.set()
```

```python
import datetime as dt
```

```python
# Generate some data
TODAY = dt.date.today()
SEVEN_DAYS = dt.timedelta(days=7)
```

```python
print( TODAY, TODAY + SEVEN_DAYS)
```

```python
pd.date_range(start=TODAY, end=TODAY + SEVEN_DAYS)
```

```python
data = {'tiempo' : pd.date_range(start=TODAY, end=TODAY + SEVEN_DAYS), 'demanda': [31, 40, 46, 37, 35, 47, 54, 42]}
data
df = pd.DataFrame(data)
df
df = df.set_index('tiempo')
```

```python
s_dec_multiplicative = seasonal_decompose(df[["demanda"]], model = "multiplicative", period = 4)
fig = s_dec_multiplicative.plot();
fig.set_size_inches((16, 9))
# Tight layout to realign things
fig.tight_layout()
plt.show()
```

```python
df[['demanda']].iloc[:].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

```python
df['SETmul12'] = ExponentialSmoothing(df['demanda'],trend='mul',seasonal='mul',seasonal_periods=4, freq = 'D').fit().fittedvalues
df.head()
```

```python
df[['demanda', 'SETmul12' ]].iloc[:].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

```python
objeto = ExponentialSmoothing(df['demanda'],trend='mul',seasonal='mul',seasonal_periods=4, freq = 'D').fit()
objeto.summary()
```

```python
forecast_values = objeto.forecast(steps=len(df))
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(df['demanda'], forecast_values)
print(f"Mean Absolute Error (MAE): {mae}")

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(df['demanda'], forecast_values)
print(f"Mean Squared Error (MSE): {mse}")

rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
model_mul = ExponentialSmoothing(df['demanda'],trend='mul',seasonal='mul',seasonal_periods=4, freq='D').fit()
forecast_mul = model_mul.forecast(8)
```

```python
import matplotlib.pyplot as plt

df['demanda'].plot(figsize=(12,8))
forecast_mul.plot(label='Forecast multiplicative model');
plt.legend(loc='upper left')
```
